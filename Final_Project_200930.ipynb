```python
#Q1

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from astropy.io import fits
import seaborn as sns
a= pd.read_csv('Pleiades_Cluster.csv')
d=1000/a.plx
d=d*3.26156
M=(5*np.log(d)-5)
sns.set_theme()
sns.displot(data=d,color='y')
mean_dis=np.sum(d)/1000

print("Mean Distance=",mean_dis)
```

```python
#Q2
from astropy import constants as const
from astropy.constants import iau2015 as const15
from astropy import units as u 
L=1000*np.exp((4.77-M)/2.5)
T=5601/(np.power(2/3,a.bp_rp+0.4))

R=(np.power(0.5,L/const15.L_sun))/(np.power(2,T/5800))

plt.figure(figsize=(10,6))
plt.plot(R,'r')
plt.plot(a.radius_val,'g')
plt.title('R vs Radius',size=20)
plt.grid()
plt.show()

plt.figure(figsize=(10,6))
plt.plot(L,'r')
plt.plot(a.lum_val,'g')
plt.title('L vs L0',size=20)
plt.grid()
plt.show()

plt.figure(figsize=(10,6))
plt.plot(T,'r')
plt.plot(a.teff_val,'g')
plt.title('L vs L0',size=20)
plt.grid()
plt.show()

L
```

```python
#Q4
x= a.bp_rp
y= M
plt.figure(figsize=(14,10),facecolor='yellow')
ax=plt.subplot(111,facecolor='black')

#plt.gca().invert_yaxis()
plt.grid()
#ax.set_xlim(-0.5,3.5)
#ax.set_ylim(15,-3)
col= []

for i in range (0, len(x)):
    if x[i]<2.5:
        col.append("cyan")
    else:
        col.append("yellow")

for i in range (0, len(x)):
    plt.scatter(x[i],y[i],c=col[i],s=10, linewidth=0)
plt.scatter(x,y,s=2,color="white")
plt.xlabel("Color",size=18)
plt.ylabel("Absolute Magnitude",size=18)
plt.title("HR Diagram(Pleiades Cluster)", size=20)
plt.show()
```
